## 3. Data Processing <a class="anchor" id="chapter3"></a>
---
In order to run analysis on the data, certain steps will be taken to clean and merge the data.
- Convert date and time data in the respective data sets into datetime objects.
- Checking and removal of duplicate entries.
- Treatement of null values identified.
- Train data set has a scenario in which if more than 50 mosquitos were caught in a trap, the trap will contribute to 2 rows of data, with the first row having a cap of 50 mosquitos and the remaining mosquitos flowing to the subsequent row. This actually places an emphasis of places with a high mosquito count and might cause bias in the data. As such, such scenarios will be identified and combined into a single row of data.
- Mosquitos take about a week from hatching as larvaes to reaching full adulthood [[source](https://www.cdc.gov/dengue/resources/factsheets/mosquitolifecyclefinal.pdf)], and the effect of spraying it to kill larvae that has hatched from eggs, and adult mosquitos [[source](https://www.cdc.gov/westnile/vectorcontrol/aerial-spraying.html#:~:text=Spraying%20larvicides%20kills%20mosquito%20larvae,permanently%20get%20rid%20of%20them.)]. Given this information, we will merge the date of last sprayed for each location into the main data set, to evaluate the effectiveness of sprays on the presence of the WNV.
- Weather conditions are also said to affect the spread of WNV as certain weather conditions make it conducive for mosquitos to breed and reproduce [[source](https://www.forbes.com/sites/brucelee/2018/09/16/west-nile-virus-how-climate-change-may-be-contributing-to-its-spread/?sh=797f29e93784)]. However, the presence of WNV detection in the master data set is based on when the trap was collected. Given that the life cycle of mosquitos take about a week for it to reach from larvae to adulthood, the weather conditions will be aggregated based on the past 7 days before being added as features to the main data set. Analysis between the effect of weather on the presence of WNV can then be better evaluated.

### 3.4 Combining Spray Data <a class="anchor" id="chapter3_4"></a>

Location based on the `Latitude` and `Longitude` will be used to combine the data. According to this [source](https://www.mosquito.org/page/faq#How%20fast%20can%20mosquitoes%20fly?), mosquitos have a flight range of 1-3 miles. In addition,  For each point in the main data set, we will create a new feature indicating if:
- `1` for there has been a spray within a 5 mile radius within the last 14 days
- `0` there has been no spray within a 5 mile radius within the last 14 days

```python
# Create geo based dataframes as distance will be used for measurements
def geo_df(df):
    geo = [Point(xy) for xy in zip(df.Longitude, df.Latitude)] # Defining geometry points
    geo_df = gpd.GeoDataFrame(df, geometry = geo)
    
    geo_df['coord'] = list(zip(geo_df.Latitude, geo_df.Longitude)) # Creating coordinate feature
    
    return geo_df

geo_train = geo_df(train)
geo_spray = geo_df(spray)
```

```python
# Function for creating marker if area within 3 miles has been sprayed in the last 14 days
def spray_marker(data, days):
    
    data['spray_14'] = 0 # By default, area will be marked as not been sprayed

    for index, row in data.iterrows(): # Iterate though all data columns
        # Creates set of days for defined period
        date_range = set(row['Date'] - datetime.timedelta(days=x) for x in range(1, days+1))
        # Finding the dates within the period that intersect with record of spray dates
        intersect = set(pd.to_datetime(geo_spray['Date'].unique())).intersection(date_range)

        # If there is an intersection, meaning there is a date within the last 14 days that a spray has been conducted
        if len(intersect) != 0:
            # Creating subset of sprays based on intersecting dates
            spray_subset = geo_spray[geo_spray['Date'].isin(intersect)]
            # Geometric points of sprays
            spray_union = spray_subset.geometry.unary_union
            # Geometric points of traps
            trap_point = row.geometry
            # Getting nearest spray point to trap
            queried_coord, nearest_coord = nearest_points(trap_point, spray_union)
            # Defining the nearest spray point in the last 14 days
            spray_coord = nearest_coord.y, nearest_coord.x
            # Measuring great-circle distance between the coordinates
            dist = geodesic(data['coord'][index], spray_coord).miles
            
            # Editing marker if distance is less than or equal to 5
            if dist < 5:
                data['spray_14'][index] = 1
            else:
                pass
        
        else:
            pass
        
# Running function for mosquito period of 2 weeks/14days
spray_marker(geo_train, 14)
```

### 3.5 Combining Weather Data <a class="anchor" id="chapter3_5"></a>

Weather also has an impact on the spread of mosquitos as it can help to create conducive conditions for their breeding. Given that we are mostly concerned with the temperature and humidity, we will mainly be focusing on the data of these 2 points. Humidity can be measured through Relative Humidity through the [Magnus Approximation](https://en.wikipedia.org/wiki/Dew_point) which is calculated from the temperature and the dewpoint, both of which are available in the weather data sets. The weather data set is also organised into measurements between two stations. We will further explore the differences between these two stations.

```python
# Investigating data types
print(test.dtypes)
```

It is observed that the average temperature recorded daily `TAvg` and the total precipitation `PrecipTotal` are object types instead of numerical types. Further investigation and correction of these data types should be done before proceeding.

```python
print(weather.Tavg.unique())
print('-------------------------')
print(weather.PrecipTotal.unique())
```

For average temperatures, data type is object due to value `M`, which is a missing value according to the data dictionary. Given that `Tavg` is the average temperature of the day, we will fill missing values with the midpoint of `Tmax` and `Tmin`.

```python
# Iterate through all rows
for index, row in weather.iterrows():
    if weather['Tavg'][index] == 'M': # If missing value
        weather['Tavg'][index] = (weather['Tmax'][index] + weather['Tmin'][index])/2 # Replace with average of Tmax and Tmin
    else:
        pass

weather['Tavg'] = weather['Tavg'].astype('float64') # Convert to float data type
```

For total precipitation, `T` indicates a trace amount of precipitation which is insignificant. For missing values denoted by `M`, we will replace with the recording from the other station.

```python
# Check that we can use the other station's precipitation recording to replace missing values
display(weather[weather['PrecipTotal']=='M'])
```

```python
weather['PrecipTotal'].replace('  T', 0, inplace=True) # Replace trace values with 0

for index, row in weather.iterrows(): # Iterate through the rows
    if weather['PrecipTotal'][index] == 'M': # If missing value
        weather['PrecipTotal'][index] = weather['PrecipTotal'][index-1] # Replace with prevoius row with data from Station 1
    else:
        pass
    
weather['PrecipTotal'] = weather['PrecipTotal'].astype('float64') # Converto to float data type
```

```python
# Check that dtypes for Tavg and PrecipTotal have successfully been converter
print(weather.dtypes)
```

The weather data also has readings for each date from 2 different stations. Exploration into the readings between these 2 stations will be explored.

```python
# Extracting important features required from weather data
weather_extract = weather[['Station', 'Date', 'Tavg', 'PrecipTotal', 'DewPoint']]
weather_extract.head()
```

```python
# Plotting difference between average temperatures from the 2 stations
plt.figure(figsize=(15,9))
sns.lineplot(y = 'Tavg', x = 'Date', data = weather_extract, hue = 'Station', palette='pastel')
plt.title('Average temperatures recorded by Station 1 and 2.',fontsize=14);
```

```python
# Plotting difference between total precipitation from the 2 stations
plt.figure(figsize=(15,9))
sns.lineplot(y = 'PrecipTotal', x = 'Date', data = weather_extract, hue = 'Station', palette='pastel')
plt.title('Total Precipitation recorded by Station 1 and 2.',fontsize=14);
```

Average temperatures across the 2 stations are very similar. For total precipitation, while the trends across the two stations are similar, there are certain days where 1 station shows significant differences in terms of the level of precipitation recorded. Given that trends from both stations are similar, we will average the values to obtain an average reading per day.

```python
# Average out the values between two station to get one value per date
weather_extract = weather_extract.set_index('Date').resample('D').mean().drop('Station',axis=1)
display(weather_extract.head())
```

Weather has an effect on the growth and spread of the mosquitos. Therefore, we are less concerned about the weather on the day on the trap as compared to the weather in the past week, which would give rise to either favourable or unfavourable conditions for their growth. The values would he ammended to the average of the values in the past 7 days to use in the analysis.

```python
# Overwrite values with average values of past week
weather_extract = weather_extract.rolling(7).mean().dropna()
display(weather_extract.head())
```

From these datapoints, we are able to compute the relative humidity based on the [Magnus Approximation](https://en.wikipedia.org/wiki/Dew_point).

```python
# Compute Relative Humidity
Tavg_C = ((weather_extract['Tavg'] - 32) * 5 / 9)
DewPoint_C = ((weather_extract['DewPoint'] - 32) * 5 / 9)
VapPress_Sat = np.exp((17.625 * Tavg_C) / (Tavg_C + 243.04))
VapPress_Act = np.exp((17.625 * DewPoint_C) / (DewPoint_C + 243.04))
R_Humidity = (VapPress_Act / VapPress_Sat) * 100

weather_extract['R_Humid'] = R_Humidity
```

```python
# Join weather data with main data
train_processed = pd.merge(left = geo_train, right = weather_extract, on='Date')
test_processed = pd.merge(left = test, right = weather_extract, on = 'Date')
```
